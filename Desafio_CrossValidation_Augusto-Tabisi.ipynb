Probemos con el metodo de Bosques Aleatorios:

```python
from sklearn.ensemble import RandomForestRegressor
ran = RandomForestRegressor(n_estimators=90)
```

Vemos que el resultado mejoro muchisimo con respecto a la regresion lineal, por ende vamos a evaluar con metricas este modelo de entrenamiento con bosques aleatorios

Train mayor over.

```python
y_pred=ran.predict(x_test)
```

```python
from sklearn import metrics
print('R2:',metrics.r2_score(y_test, y_pred))
print('MSE:',metrics.mean_squared_error(y_test, y_pred))
```

Vemos que el R2 muestra casi un 90% por lo tanto podriamos decir que las variables aplicadas en el metodo de bosques aleatorios explican de manera aceptable el precio de los alquileres.

```python
from sklearn.model_selection import cross_validate


model = LinearRegression()
scores1 = cross_validate(model, x_train, y_train, cv=5, scoring='neg_mean_squared_error',
                        return_train_score=True)


print('Cross-Validation in Validation set MSE:', -np.mean(scores1['test_score']))

print('Cross-Validation in train set MSE:', -np.mean(scores1['train_score']))
```

**La diferencia entre los MSE es bastante pequeña**. Dado que la diferencia es minima, el modelo está generalizando bien no se nota overfitting (sobreajuste).

```python
-np.mean(scores1['test_score']) - -np.mean(scores1['train_score'])
```

```python
import pandas as pd
import numpy as np
import requests
```

```python
census_api_key = ''
variable = 'B01003_001E'
year = '2020'
url1 = 'https://api.census.gov/data/{0}/acs/acs5?get=NAME,{1}&for=us:*&key={2}'\
  .format(year, variable, census_api_key)
response = requests.request("GET", url1)
response.status_code
```

```python
response
```

```python
response.text
```

```python
df_USA = pd.DataFrame(response.json()[1:], columns=response.json()[0]).rename(columns={'B01001_001E' :'Cantidad de poblacion'})
df_USA
```

```python
census_api_key = ''
variable = 'B01001_001E'
year = '2020'
# Data por estado
url2 = 'https://api.census.gov/data/{0}/acs/acs5?get=NAME,{1}&for=state:*&key={2}' .format(year, variable, census_api_key)
response = requests.request("GET", url2)
response.status_code
```

```python
df_estados = pd.DataFrame(response.json()[1:], columns=response.json()[0]).rename(columns={'index': (0),'NAME': 'Estado','B01001_001E' :'Cantidad de habitantes', 'state' : 'Numero de Estado'})
df_estados
```

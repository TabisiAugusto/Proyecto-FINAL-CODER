Traemos las librerias necesarias:

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
fig, ax = plt.subplots()
ax.boxplot( x = df['price'])
fig.set_size_inches(15, 10)
ax.set_title('Boxplot de los Precios', loc='center', fontdict={'fontsize':14, 'fontweight':'bold', 'color':'tab:red'})
plt.show()
```

Hay problemas con el precio, como podemso ver la presencia de un outsider bastante pronunciado genera que no podamos analizar ni siquiera la forma de la caja porque al estar tan alejado la escala no nos permite verla.

```python
fig, ax = plt.subplots()
ax.boxplot( x = df['sqfeet'])
fig.set_size_inches(15, 10)
ax.set_title('Boxplot de los Pies cuadrados', loc='center', fontdict={'fontsize':14, 'fontweight':'bold', 'color':'tab:red'})
plt.show()
```

Lo mismo que en el caso anterior, vemos la presencia de outsiders bastante alejados.

*Viendo los valores maximos de pies cuadrados y precios podemos decir que hay valores atipicos. El criterio en el que me baso es ver el valor maximo de ambas columnas, el cual si vemos hasta el 75% de la muestra el precio llega hasta $1395 y luego el maximo se encuentra alojado en un numero con nueve ceros luego de la coma, lo cual no tiene logica a pagar por un alquiler a menos que la persona que lo busque sea millonario. Por el lado de los pies cuadrados podemos ver que el 75% de la muestra llega hasta 1150 pies, donde luego nuevamente el maximo se aloja en un numero con 6 ceros luego de la coma, provocando otra ves problemas para el analisis posterior. Por ende tomo los siguientes recaudos, atado a posteriores modificaciones de ser necesario, para poder empezar con los graficos.Tambien tenemos algunas preguntas como por ejemplo, quien alquilaria una casa con 75 baños, 35 baños o 25 baños. Suponiendo que la persona sea adinerada y se pudiera dar el lujo de alquilar alguna de las casas disponibles, el tener mas de 7 baños ya seria una casa totalmente excesiva para que viva cualquier familia. Por ende vamos a tener que modificar algunas cosas. Lo mismo sucede cuando hablamos de las camas, vemos que hay registros con mas de 1000 camas disponibles, una locura, a menos que se alquile un hotel entero.*

```python
limpieza1 = ((df["sqfeet"]<120) | (df["sqfeet"]>3000) | (df["price"]<100) | (df["price"]>4000))
df = df[~limpieza1]
```

```python
df.describe().T
```

Volviendo a realizar los graficos:

```python
fig, ax = plt.subplots()
ax.boxplot( x = df['price'])
fig.set_size_inches(15, 10)
ax.set_title('Boxplot de los Precios', loc='center', fontdict={'fontsize':14, 'fontweight':'bold', 'color':'tab:red'})
plt.show()
```

```python
fig, ax = plt.subplots()
ax.boxplot( x = df['sqfeet'])
fig.set_size_inches(15, 10)
ax.set_title('Box plot pies cuadrados', loc='center', fontdict={'fontsize':14, 'fontweight':'bold', 'color':'tab:red'})
plt.show()
```

Para comenzar con las visualizaciones, utilizaremos seaborn para conocer el tipo de alquileres que mayor presencia tienen el dataset, asi como una discriminacion por tipo de alquiler. Se utiliza un grafico que realiza un conteo.

```python
sns.countplot(df["type"])
fig = plt.gcf()
fig.set_size_inches(20,15)
plt.title('Cantidad de alquileres disponibles diferenciados por tipo', size= (15))
```

Podemos ver la predominancia del alquiler del tipo departamento. Esto puede estar relacionado con el incremento de los ultimos años de las construcciones de edificios en las ciudades, asociado con la creciente poblacion. Seguido a los departamentos, estan las casas y por ultimo las casas residenciales, las cuales pueden ser encontradas en los barrios privados en USA. 

Parecido al grafico anterior, pero ahora discriminando por estados, veremos la cantidad de alquileres presentes en cada uno, con el fin de analizar mas a detalle el numero de alquileres presente en cada uno de los estados. Para esto se utilizar un grafico de conteo

```python
sns.countplot(df["state"])
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.title('Alquileres disponibles por estados', size=(15))
```

Observando el grafico anterior, podemos ver que en tres estados sobrepasan facilmente la barrera de los 25000 alquileres disponibles. Los cuales son el estado de California, Florida y Texas. Seguidos por Carolina del Norte y Michigan, completando asi el top 5 de Estados con mayor disponibilidad de alquileres.

Siguiendo con lo anterior se presenta un grafico de dispersion para analizar los precios de alquileres presente en los diferentes estados y asi dar aun mas detalle sobre lo que ocurre en cada uno de ellos.

```python
fig1, ax = plt.subplots()
ax.scatter(df['state'], df['price'], alpha=0.5)
fig1.set_size_inches(20, 10)
ax.grid(axis = 'y', color = 'black', linestyle = 'dashed')
ax.set_title('Distribucion de los precios en base a los alquileres disponibles por Estado', loc='center', fontdict={'fontsize':14, 'fontweight':'bold', 'color':'tab:red'})
plt.show()
```

Con respecto a los precios que se observan en los diferentes estados, es posible apreciar que California es el estado con la mayor cantidad de precios disponibles, lo cual a su vez puede estar relacionado con que es el estado con mayor disponibilidad de alquileres.

En este grafico utilizaremos un histograma para ver la forma de la distribucion de los precios, lo cual para comenzar con el analisis nos ayudara a saber en donde estamos "parados" con respecto a los precios que prevalecen en los alquileres del dataset presente.



```python
sns.histplot(df["price"], color='red')
fig = plt.gcf()
fig.set_size_inches(15,10)
plt.title('Distribucion de precios', size=(15))
```

Viendo el grafico anterior podemos observar la presencia de una distribucion sesgada hacia la izquierda, donde la maxima frecuencia esta alojada entre un valor de 750 a 1000 aproximadamente. 

A continuacion utilizamos un diagrama de distribucion, con la finalidad de dar en cuenta sobre la distribucion de los pies cuadrados en el dataset para luego analizar si existe cierta correlacion con el precio de los alquileres.

```python
sns.distplot(df["sqfeet"])
fig = plt.gcf()
fig.set_size_inches(15,10)
plt.title('Distribucion de Pies Cuadrados', size=(15))
```

Analizando el grafico anterior encontramos una distribucion sesgada hacia la izquierda, con una concentracion de los pies cuadrado en el rango de 500 a 1500 pies cuadrados aproximadamente.

Se presenta un diagrama de dispersion entre precios y pies cuadrados para analizar si se distribuyen de manera conjunta o si no hay un patron claro de ver.

```python
fig2, ax = plt.subplots()
ax.scatter(df['sqfeet'], df['price'])
fig2.set_size_inches(20, 10)
ax.set_xlabel('Pies cuadrados')
ax.set_ylabel('Precio')
ax.set_title('Relacion Precio/Pies Cuadrados', loc='center', fontdict={'fontsize':14, 'fontweight':'bold', 'color':'tab:red'})
plt.show()
```

Podemos observar que no hay una relacion definida entre el precio de los alquileres con el tamaño en pies cuadrados de los mismos. Por lo tanto no estamos en condiciones de afirmar ninguna relacion entre precio y tamaño.
